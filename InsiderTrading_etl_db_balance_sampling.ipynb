```
import pandas as pd
from imblearn.over_sampling import SMOTE

insider_df = pd.read_csv("./drive/My Drive/DatosInsider/DWH_tables/InsiderDataMart.csv", index_col=0)
insider_df.drop(['period_y','user'], axis=1,inplace=True)
insider_df['gender']= insider_df['gender'].apply(lambda x: 1 if x=='male' else 0)
insider_df.rename(columns = {'period_x':'period'}, inplace = True) 
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
from sklearn import model_selection

insider_OOT = insider_df[insider_df['period'] > 201102]
insider_IT = insider_df[insider_df['period'] < 201103]
insider_Y= insider_IT['isInsider']
insider_X= insider_IT.drop(['isInsider'], axis=1)
ins_train_x,  ins_test_x,ins_train_y ,ins_test_y = model_selection.train_test_split(insider_X, insider_Y, test_size=0.3,random_state=2020,stratify=insider_Y )
```

```
import os
path="./drive/My Drive/DatosInsider/DWH_tables/"

ins_train_x.to_csv(os.path.join(path, "ins_train_x.csv"), index=False)
ins_test_x.to_csv(os.path.join(path, "ins_test_x.csv"), index=False)
ins_train_y.to_csv(os.path.join(path, "ins_train_y.csv"), index=False)
ins_test_y.to_csv(os.path.join(path, "ins_test_y.csv"), index=False)
```

```
from imblearn.over_sampling import SMOTE

ins_train_x_resampled, ins_train_y_resampled = SMOTE(random_state=2020).fit_resample(ins_train_x, ins_train_y)

```

```
x_cols= ins_train_x.columns
ins_train_x_resampled_df=pd.DataFrame(ins_train_x_resampled,columns=x_cols)
```

```
#'y_cols= ins_train_y.columns
ins_train_y_resampled_df=pd.DataFrame(ins_train_y_resampled,columns=['isInsider'])
```

```
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression


estimator = LogisticRegression(max_iter=10000)
selector = RFECV(estimator, step=1, cv=5)
selector = selector.fit(ins_train_x_resampled, ins_train_y_resampled)
selector.support_

```
